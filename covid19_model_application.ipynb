This is the key names for the columns and what question they are referring to

time         | Timestamp                                                                                                                                                                             
gender       | What is your gender?                                                                                                                                                                  
ls_before    | Generally, how would you say you felt about life BEFORE the impacts the pandemic and social distancing?                                                                               
ls_after     | Generally, how would you say you felt about life DURING the impacts the pandemic and social distancing?                                                                               
affected     | Did anyone you know (including yourself) get seriously affected by the COVID-19 Virus in which it brought great stress to your life?                                                  
lost_job     | Did you lose your job during this pandemic? (And you are still unemployed)                                                                                                            
work_method  | If you worked through this pandemic, what was your main method of work?  Select N/A if you did not work                                                                               
residents    | Who did you live with during this pandemic?                                                                                                                                            
outdoor_time | On average, how much time do you spend outside per day?                                                                                                                                
exercise_time| On average, how much time do you spend exercising per day?                                                                                                                             
sleep_time   | Using this scale, where one bubble is one hour, around how many hours of sleep do you get per night? If you sleep over 10 hours a night, select 10.                                    
importance   | Which areas of life are highly important to you?                                                                                                                                       
hobbies      |What were your main hobbies during the pandemic? (Hobbies you spend around 3-4+ hours doing a week)                                                                                    
news         |Do you keep up to date with current events and news?                                                                                                                                   

sample_bias  |This question is completely optional and will only be used to more accurately understand any sampling bias at play. How did you hear of this survey? (i.e. Tom's instagram story)     

```python
covid_df.columns = ['time', 'gender','ls_before','ls_after','affected','lost_job','work_method','residents','outdoor_time','exercise_time','sleep_time','importance','hobbies','news','sample_bias']
covid_df.drop(['sample_bias'], axis=1, inplace=True)
covid_df.drop(['time'], axis=1, inplace=True)
```

```python
covid_df.isnull().sum()
```

```python
covid_df['work_method'] = covid_df.work_method.fillna('Does_not_work')
covid_df['lost_job'] = covid_df.lost_job.fillna('No')
```

```python
covid_df = (covid_df[~covid_df['importance'].isnull()]).reset_index(drop=True)
```

```python
covid_df.isnull().sum()
```

```python
for i in covid_df.columns:
    print(str(i) + str(covid_df[i].value_counts()))
```

```python
#Converting variables to integer values where the description is listed out in the readme file
covid_df['affected'] = (covid_df['affected'] == 'Yes').astype(int)
covid_df['lost_job'] = covid_df['lost_job'].isin(['Yes','Furloughed']).astype(int)
covid_df = covid_df[covid_df['gender'].isin(['Male','Female'])].reset_index(drop=True)
covid_df['gender'] = covid_df['gender'].isin(['Male']).astype(int)
```

```python
covid_df['work_method'].value_counts()
```

```python
#Converting >2 value possibility columns to new separate columns
#Travel_Method
covid_df['method_travel'] = covid_df['work_method'].str.contains('Travel').astype(int)
covid_df['method_home'] = covid_df['work_method'].str.contains('home').astype(int)
covid_df['method_no_work'] = covid_df['work_method'].str.contains('not').astype(int)
```

```python
#Living_condition
covid_df['live_alone'] = covid_df['residents'].str.contains('Alone').astype(int)
covid_df['live_parents_siblings'] = covid_df['residents'].str.contains('Siblings').astype(int)
covid_df['live_friends_roommates'] = covid_df['residents'].str.contains('Friends/Roommates').astype(int)
covid_df['live_children'] = covid_df['residents'].str.contains('Children').astype(int)
covid_df['live_partner'] = covid_df['residents'].str.contains('Partner').astype(int)
```

```python
#Living_condition - double check to see values are good
print(covid_df['live_alone'].value_counts())
print(covid_df['live_parents_siblings'].value_counts())
print(covid_df['live_friends_roommates'].value_counts())
print(covid_df['live_children'].value_counts())
print(covid_df['live_partner'].value_counts())
```

```python
#Convert these time ranges into a more workable time series. Will scale everything down to hours and use the average of the range for the value.

# N/A - You do not go outdoors   --> 0 
# Less than 30 minutes           --> 0.25 
# Between 30 minutes to 1 hour   --> 0.75
# Between 1 hour and 2 hours     --> 1.5  
# More than 2 hours              --> 2.75   -Just an assumption  

covid_df['outdoor_time'] = np.where((covid_df.outdoor_time == 'N/A - You do not go outdoors'),0,covid_df.outdoor_time)
covid_df['outdoor_time'] = np.where((covid_df.outdoor_time == 'Less than 30 minutes'),0.25,covid_df.outdoor_time)
covid_df['outdoor_time'] = np.where((covid_df.outdoor_time == 'Between 30 minutes to 1 hour'),0.75,covid_df.outdoor_time)
covid_df['outdoor_time'] = np.where((covid_df.outdoor_time == 'Between 1 hour and 2 hours'),1.5,covid_df.outdoor_time)
covid_df['outdoor_time'] = np.where((covid_df.outdoor_time == 'More than 2 hours'),2.75,covid_df.outdoor_time)
```

```python
#Convert these time ranges into a more workable time series. Will scale everything down to hours and use the average of the range for the value.

# Name: exercise_time
# N/A - You do not spend time exercising  --> 0
# Less than 30 minutes                    --> .25  
# Between 30 minutes to 1 hour            --> .75
# More than 1 hour                        --> 1.25  

covid_df['exercise_time'] = np.where((covid_df.exercise_time == 'N/A - You do not spend time exercising'),0,covid_df.exercise_time)
covid_df['exercise_time'] = np.where((covid_df.exercise_time == 'Less than 30 minutes'),0.25,covid_df.exercise_time)
covid_df['exercise_time'] = np.where((covid_df.exercise_time == 'Between 30 minutes to 1 hour'),0.75,covid_df.exercise_time)
covid_df['exercise_time'] = np.where((covid_df.exercise_time == 'More than 1 hour'),1.25,covid_df.exercise_time)
```

```python
covid_df['exercise_time'].value_counts()
```

```python
#importance 
covid_df['imp_relationships'] = covid_df['importance'].str.contains('Relationships').astype(int)
covid_df['imp_career'] = covid_df['importance'].str.contains('Career').astype(int)
covid_df['imp_social_justice'] = covid_df['importance'].str.contains('Justice').astype(int)
covid_df['imp_religion_faith'] = covid_df['importance'].str.contains('Religion').astype(int)
```

```python
#hobbies
covid_df['hobby_videos'] = covid_df['hobbies'].str.contains('Movies').astype(int)
covid_df['hobby_reading'] = covid_df['hobbies'].str.contains('Reading').astype(int)
covid_df['hobby_outdoor'] = covid_df['hobbies'].str.contains('Outdoor').astype(int)
covid_df['hobby_instrument'] = covid_df['hobbies'].str.contains('Instrument').astype(int)
covid_df['hobby_reading'] = covid_df['hobbies'].str.contains('Reading').astype(int)
```

```python
# #News
# Yes       --> 1
# Sometimes --> 0.5           
# No        --> 0     
covid_df['news'] = np.where((covid_df.news == 'No'),0,covid_df.news)
covid_df['news'] = np.where((covid_df.news == 'Sometimes'),0.5,covid_df.news)
covid_df['news'] = np.where((covid_df.news == 'Yes'),1,covid_df.news)
```
